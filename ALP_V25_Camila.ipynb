```python
#Calculadora de IMC

peso = float(input("Digite seu peso: "))
altura = float(input("Digite sua altura: "))
imc = peso / (altura * altura)

print("Seu IMC é: ", imc)
```

##Estrutura Condicional

if - se

elif - senão se

else - senão

```python
nota = float(input("Digite sua nota: "))

if (nota >= 7):
  print("Aprovado")
elif (nota >= 5):
  print("Recuperação")
else:
  print("Reprovado")
```

4 - Crie um algoritmo que receba uma idade do usuário, e depois verifique em qual feixa etária está à idade:

até 12 - Você é uma criança!

de 13 à 18 - Você é um Adolescente!

de 19 à 60 - Você é Adulto!

acima de 60 - Você é idoso!

```python
idade = int(input("Digite sua idade: "))

if idade <= 12:
  print ("Você é uma criança!")
elif idade < 18:
  print ("Você é um adolescente!")
elif idade <= 60:
  print ("Você é um adulto!")
else:
  print ("Você é um idoso!")
```
